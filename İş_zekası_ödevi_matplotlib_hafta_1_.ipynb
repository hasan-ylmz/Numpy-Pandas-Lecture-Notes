Python dictionary kullanarak bir data frame oluşturun. DataFrame içerisinde 0-10000 arasında rastgele tamsayılardan oluşan sayilar isimli bir feature olmalıdır. Daha sonra list compherension kullanarak durum isimli bir feature oluşturun ve eğer değerler 5000 ve daha az ise düşük daha yüksek ise yüksek olacak şekilde değerlerini oluşturun. Bir parametre alan ve hesaplama isimli fonksiyon tanımlayın. Fonksiyon girilen sayısal değerin yarısının 5 fazlasını geri döndürsün. Daha sonra hesaplanan adında yeni bir feature oluşturun ve sayilar isimli feature değerlerinden yararlanarak ve apply kullanarak hesaplama fonskyionunu kullanıp hesaplatın. Son olarak x ekseninde hesaplanan, y ekseninde de sayılar featurelerinin olduğu bir grafik (çizgi grafiği) çizdirin.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'sayılar' :np.random.randint(0,10000,size=554)})
```

```python
df['durum'] =['düşük' if x<=5000 else 'yüksek' for x in df['sayılar']]
```

```python
def hesaplama(x):
    return x/2+5
```

```python
df['hesaplanan'] =df['sayılar'].apply(hesaplama)
```

```python
plt.figure(figsize=(5,5))
plt.plot(df['hesaplanan'] ,df['sayılar'])
plt.xlabel("hesaplanan")
plt.ylabel("Sayılar")
plt.title("Hesaplanan ve sayılar")
plt.show()
```
